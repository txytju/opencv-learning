- opencv使用BGR作为默认的数据格式，matplotlib使用RGB作为默认的数据格式，因此使用opencv导入的图片如果想用matplolib正常显示需要将数据格式转换成RGB
- 同理，将图片转换为gray之后，如果想要使用matplotlib正确绘制出来，也需要指定数据的格式
- 上述两个情况发生的原因都是
    - 图片是以不同维度的矩阵存在的，使用这些矩阵时，需要指定不同维度的意义才能正确显示

```python
import cv2  
import numpy as np  
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("landscape2.png")
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)

# the convert from BGR to RGB
print "the 1st channel of img and the last channel of RGB_img is the same : ", (img[:,:,0] == RGB_img[:,:,2]).all()

plt.imshow(RGB_img)
plt.title('my picture')
plt.show()

# after convert to gray, only one dim is kept
print "the shape of gray image is ", gray.shape
```

```python
plt.imshow(gray, "gray") # need to specify that the iamge is gray image
plt.title('my picture')
plt.show()
```
